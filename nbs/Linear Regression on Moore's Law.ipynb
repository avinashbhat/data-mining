```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
```

```python
non_decimal = re.compile(r'[^\d]+')
for line in open('../data/linear_regression_data/moore.csv'):
    r = line.split('\t')
    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
Y = np.log(Y)
plt.scatter(X, Y)
plt.show()
```

```python
denom = X.dot(X) - X.mean() * X.sum()
```

```python
a = (X.dot(Y)- Y.mean() * X.sum()) / denom
```

```python
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y)) / denom
```

```python
y_hat = a*X + b
```

```python
plt.scatter(X, Y)
plt.plot(X, y_hat)
plt.show()
```

```python
d1 = Y - y_hat
d2 = Y - Y.mean()
r_squared = 1 - d1.dot(d1) / d2.dot(d2)
print(r_squared)
```
